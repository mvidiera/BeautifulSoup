```python
import bs4
import requests
import time
import random as ran
import sys
import pandas as pd
```

```python
url = 'https://www.verizon.com/business/products/plans/'
```

```python
source = requests.get(url).text
soup = bs4.BeautifulSoup(source,'html.parser')
print(soup)
```

```python
anchors= soup.find_all('a')
all_links= set()

for i in anchors:
    if(i.get('href')!= "#"):
        link= "https://www.verizon.com" +i.get('href')
        all_links.add(link)
        print(link)
```

```python
# program above will fetch the hyperlinks which are suffix of vz.com
# to get the children of a class
# consider a class: gnav20-main

classcontent= soup.find(id='main')
print(classcontent)
```

```python
print(classcontent.children)
```

```python
# to print all the contents like div, a, etc 
print(classcontent.contents)
```

```python
# toprintthe contents 

for elements in classcontent.contents:
    print(elements)
```

```python
for elements in classcontent.contents:
    print(elements)
```

#### Difference btw children and contents

- .contents: list of tag object. contents are stored in memory. Contents are iterable
- .children:generator of tag object. Children are not stored in memory but canbe fetched using for loop or next()
- .children is faster for huge pages. 

```python
for elements in classcontent.strings:
    print(elements)
    
# within this class, strings are printed here
```

```python
for elements in classcontent.stripped_strings:
    print(elements)
    
#all strings are displayed one after other 
```

```python
# to find parent of any element 

print(classcontent.parent)
#parent of classcontent is <body> 
```

```python
print(classcontent.parents)

#parents print generator. as it is generator, let me apply for loop to display
```

```python
for item in classcontent.parents:
    print(item)
```

```python
for item in classcontent.parents:
    print(item.name)

#tree navigation: as shown in image above, my tag looks for its parents using .parents (plural)
# immediate parent of classcontent is body. then html (first tag)
```

#### Finding out Siblings: Next and previous sibling. 

```python
print(classcontent.next_sibling)
#this classcontent's next sibling is space. as space or new lines are also considered as sibling
#hence space is returned as output
```

```python
print(classcontent.next_sibling.next_sibling)
#
```

```python
print(classcontent.previous_sibling.previous_sibling)
```


- Though I have used previous_sibling.previous_sibling( meaning that skip 2 steps) sibling's sibling
- Yet, I got previous sibling. Why? 
- Next lines, spaces are also considered as sibling. Hence when I printed .next_sibling(once), space was returned instead of any tags.

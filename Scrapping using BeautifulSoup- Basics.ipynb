```python
# all of these are repeated
# few links here like (smartphones/apple-iphone14) are not hyperlinks
#I can not get into the link clicking these info(not hyperlink)

anchors= soup.find_all('a')
all_links= set()
#using set here because I dont want repeated links. If I use list then even repeated links are 
#added in the list

for i in anchors:
    if(i.get('href')!= "#"):#if theere are any links starting with #(in my case it isn't)
        link= "https://www.verizon.com" +i.get('href')
        all_links.add(link)
        print(link)
        
# i iterates through anchors, and drop href which has #
# now I have all href without #, and I am adding to link variable. 
# content link variable is added in set(all_links)
#links which start with verizon.com are fetched and stored in link. 
#I'm adding this link to set all_links
#print link: it has string of all links even repeated. 
# to print unique links then print(all_links) which is of set datatype
```

```python
print(all_links)
#this returns a set of all_links
```
